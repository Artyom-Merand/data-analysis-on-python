### 10. Split the data into training set and testing set using train_test_split.

```python
from sklearn.model_selection import train_test_split
```

```python
advertising.head(5)
```

```python
X = advertising[['Daily Time Spent on Site','Age','Area Income',
                 'Daily Internet Usage','Male']]
y = advertising['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### 11. Train and fit a logistic regression model on the training set.

```python
from sklearn.linear_model import LogisticRegression
```

```python
logrm = LogisticRegression()
```

```python
logrm.fit(X_train, y_train)
```

### 12. Predict values for the testing data.

```python
predictions = logrm.predict(X_test)
```

### 13. Create a classification report for the model.

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predictions))
```
